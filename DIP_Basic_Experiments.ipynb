# Upload Image

```
from google.colab import files

uploaded = files.upload()

for filename in uploaded.keys():
    print('Uploaded file:', filename)
```

# Load and Display the Uploaded Image

```
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage

# Load the uploaded image
image_path = list(uploaded.keys())[0]  # Assuming only one image uploaded
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale

# Display the image using matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title('Uploaded Image')
plt.axis('off')
plt.show()
```

# 1. Image Enhancement: Histogram Equalization
Apply Histogram Equalization:

Enhances the image contrast.

```
# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the equalized image
plt.figure(figsize=(6, 6))
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')
plt.show()
```

# **Visualising images as arrays**

```
# Convert images to NumPy arrays
image_array = np.array(image)
equalized_array = np.array(equalized_image)

# Print the shapes of the arrays (for verification)
print(f"Original Image Array Shape: {image_array.shape}")
print(f"Equalized Image Array Shape: {equalized_array.shape}")

# Print the original and equalized arrays (for verification)
print("\nOriginal Image Array:")
print(image_array)

print("\nEqualized Image Array:")
print(equalized_array)
```

# **2. Image Restoration: Gaussian Blur**
Apply Gaussian Blur:

Reduces noise and smoothens the image.

```
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(equalized_image, (5, 5), 0)

# Display the blurred image
plt.figure(figsize=(6, 6))
plt.imshow(blurred_image, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
```

# **3. Image Segmentation: Canny Edge Detection**
Apply Canny Edge Detection:

Detects edges in the image.

```
from skimage import feature

# Apply Canny edge detection
edges = feature.canny(blurred_image, sigma=1.5)

# Display the edge-detected image
plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

# **4. Image Representation and Description**
Extract Features:

Compute basic statistics or use advanced feature extraction methods.

```
# Example: Calculate mean and standard deviation of pixel intensities
mean_intensity = blurred_image.mean()
std_dev_intensity = blurred_image.std()

print(f'Mean Intensity: {mean_intensity}')
print(f'Standard Deviation of Intensity: {std_dev_intensity}')
```
